```python
import pandas as pd
import numpy as np

# Создаём грязный датасет
data = {
    'name':    ['Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Eve', 'Frank', 'bob'],
    'age':     [25, 30, None, 25, 200, 28, 35, 30],
    'salary':  [50000, 60000, 55000, 50000, 70000, None, 45000, 60000],
    'country': ['USA', 'usa', 'UK', 'USA', 'Canada', 'uk', 'Canada', 'USA'],
    'score':   ['85', '90', '78', '85', '92', '88', None, '90']
}

df_dirty = pd.DataFrame(data)
df_dirty
```

```python
print("Размер датасета:", df_dirty.shape)
print("\nТипы данных:")
print(df_dirty.dtypes)
print("\nПропущенные значения:")
print(df_dirty.isnull().sum())
print("\nДубликаты:", df_dirty.duplicated().sum())
```

```python
df_clean = df_dirty.copy()

# Заполняем age медианой
df_clean['age'] = df_clean['age'].fillna(df_clean['age'].median())

# Заполняем salary медианой
df_clean['salary'] = df_clean['salary'].fillna(df_clean['salary'].median())

# Удаляем строки где score пустой
df_clean = df_clean.dropna(subset=['score'])

print("Пропуски после очистки:")
print(df_clean.isnull().sum())
```

```python
print("До удаления дубликатов:", df_clean.shape)
print("\nДубликаты:")
print(df_clean[df_clean.duplicated(keep=False)])

df_clean = df_clean.drop_duplicates()
print("\nПосле удаления дубликатов:", df_clean.shape)
```

```python
print("До:")
print(df_clean['country'].unique())

# Приводим к нижнему регистру и убираем пробелы
df_clean['country'] = df_clean['country'].str.upper().str.strip()

print("\nПосле:")
print(df_clean['country'].unique())
```

```python
print("До:")
print(df_clean['score'].dtype)

df_clean['score'] = df_clean['score'].astype(float)

print("После:")
print(df_clean['score'].dtype)
```

```python
print("До:")
print(df_clean['age'].describe())

# Заменяем невозможные значения на NaN а потом на медиану
df_clean.loc[df_clean['age'] > 120, 'age'] = df_clean['age'].median()

print("\nПосле:")
print(df_clean['age'].describe())
```

```python
print("=== ФИНАЛЬНЫЙ ОТЧЁТ ===")
print("\nРазмер:")
print(f"До очистки:  (8, 5)")
print(f"После очистки: {df_clean.shape}")

print("\nТипы данных:")
print(df_clean.dtypes)

print("\nПропуски:")
print(df_clean.isnull().sum())

print("\nДубликаты:", df_clean.duplicated().sum())

print("\nУникальные страны:")
print(df_clean['country'].unique())

print("\nФинальная таблица:")
df_clean
```
