```python
import pandas as pd
import numpy as np
```

```python
data = {'country': ['UsA', 'usa', 'rUSSIA', 'dcd  d', 'usa'],
 'name':['aAdi', 'viren', None, 'viren', 'viren'],
       'numbs':[123, 132, 1000, 125, 132]}

datapd = pd.DataFrame(data)


datapd['country'] = datapd['country'].str.upper().str.replace('  ', '')
datapd_clean = datapd.drop_duplicates()

datapd_clean.loc[datapd_clean['numbs'] > 135, 'numbs'] = int(datapd_clean['numbs'].median())

datapd_clean = datapd_clean.copy()
datapd_clean['name'] = datapd_clean['name'].str.capitalize()

datapd_clean['name'] = datapd_clean['name'].fillna('Unknown')

datapd_clean



```
